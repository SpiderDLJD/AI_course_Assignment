## 1. My own sentence generator

### 想创造一个与和女票说话的机器人，包含哄高兴，约吃美食两种基本功能

```python
pattern = '''
哄高兴 => 复合人称 消气动作 尾部助词
复合人称 => 形容词 复合人称 | 人称
形容词 => 亲爱的 | 可爱的 | 通情达理的 | 漂漂的 | 美美哒 | 萌萌哒
人称 => 宝贝 | 宝宝 | 小月月 | 老婆 | 娘娘 | 小仙女
消气动作 => 饶了我嘛 | 别生气啦 | 我知错了
尾部助词 => 呜呜呜 | ，人家超级惨的
约饭 => 复合人称 吃饭动作 美食 语气助词
吃饭动作 => 尝尝 | 试一试 | 吃吃看
美食 => 铁锅一居焖面 | 烤羊腿 | 海底捞 | 舒芙蕾 | 烤鲈鱼 | 披萨 | 王婆大虾 | 高兴火锅 | 西班牙海鲜饭 | 大盘鸡 | 将太无二
语气助词 => 吧 | 嘛 | 怎么样嘛
'''
```

```python
import random
split_stmt = '=>'
or_split = '|'
```

```python
pattern = [line for line in pattern.split('\n') if line != '']
```

```python
rule = dict()
```

```python
def rule_generate(pattern):
    for line in pattern:
        stmt, expr = line.split(split_stmt)
        rule[stmt.strip()] = expr.split(or_split)
```

```python
rule_generate(pattern)
```

```python
def sentence_generate(rule, target):
    if target in rule:
        candidates = rule[target]
        candidate = random.choice(candidates)
        return ''.join(
        sentence_generate(rule, c) for c in candidate.split()
        )
    else:
        return target
```

```python
def generate_n_sentence(rule, target, n):
    for i in range(n):
        print(sentence_generate(rule, target))
```

```python
generate_n_sentence(rule, '哄高兴', 10)
```

```python
generate_n_sentence(rule, '约饭', 10)
```

## 2. 使用新数据源完成语言模型的训练

### 数据源采用我最喜欢的《明朝那些事儿》，采用 3-gram 模型

```python
from collections import Counter
import jieba
```

```python
file = open('D:/Ming_dynasty.txt', encoding='utf-8').read()
```

```python
file[:1000]
```

```python
def cut(string):
    return list(jieba.cut(string))
```

```python
cut_Ming = cut(file)
```

```python
len(file)
```

```python
cut_Ming_3_gram = [cut_Ming[i] + cut_Ming[i+1] + cut_Ming[i+2] for i in range(len(cut_Ming)-3)]
```

```python
cut_Ming
```

```python
cut_Ming_3_gram
```

```python
count_Ming = Counter(cut_Ming)
```

```python
count_Ming_3_gram = Counter(cut_Ming_3_gram)
```

$$Pr(sentence) = Pr(w_1 \cdot w_2 \cdots w_n) = \prod \frac{count(w_i, w_{i+1}, w_{i+2})}{count(w_i)}$$

```python
def get_1_gram(word):
    if word in count_Ming:
        return count_Ming[word]
    else:
        return count_Ming.most_common()[-1][-1]
```

```python
def get_3_gram(word):
    if word in count_Ming_3_gram:
        return count_Ming_3_gram[word]
    else:
        return count_Ming_3_gram.most_common()[-1][-1]
```

```python
def three_gram_model(string):
    cut_string = cut(string)
    probability = 1
    for i in range(len(cut_string)-2):
        word = cut_string[i]
        word_next = cut_string[i+1]
        word_next_2 = cut_string[i+2]
        pro = get_3_gram(word + word_next + word_next_2) / get_1_gram(word)
        probability *= pro
    return probability
```

```python
three_gram_model('你是真的家没曹拉格式有点儿皮啊')
```

```python
three_gram_model('想想以后的事情吧')
```

## 3. 获得最优质的语言

### 采用第一部分自己制作的“女票对话机器人”，考虑到文中的需求，需要预先设置 数据结构为：
$$[(sentence_1, probability_1), (sentence_2, probability_2), \cdots(sentence_n, probability_n)]$$
### 其中 probability_n 是通过第二部分《明朝那些事》所得到的语料库，从而计算出每句话正确的概率。
### generate_best 函数共根据规则生成 n 句随机的话，并从中选出最正确的一句。

```python
def generate_best(rule, target, n):
    sen_pro = []
    for i in range(n):
        sen_pro.append((sentence_generate(rule, target), three_gram_model(sentence_generate(rule, target))))
    sorted(sen_pro, key=lambda x: x[1], reverse=True)
    return sen_pro[0][0]
```

```python
generate_best(rule, '约饭', 10)
```

```python
generate_best(rule, '哄高兴', 10)
```
